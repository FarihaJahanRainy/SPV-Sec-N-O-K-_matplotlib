 # Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup Plot
fig,ax=plt.subplots(figsize=(5,6))

# 3. plot data
ax.plot(x,y)

# 4. customize data
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. save and show the whole figure
plt.savefig('figures/sample-plot.png')
```

# Scatter Plot and Bar Plot

```python
# Making figures with NumPy arrays,
#Create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
#plot the data
fig,ax=plt.subplots(figsize=(5,4))
ax.plot(x,x**2)# this is called as line plot
```

```python
#use some data to make a scatter
fig,ax=plt.subplots(figsize=(5,4))
ax.scatter(x,np.exp(x))# creating an exponential graph
```

```python
#another scatter plot with sin function
fig,ax=plt.subplots(figsize=(5,4))
ax.scatter(x,np.sin(x))
```

```python
## Bar Plot
# Bar plot using python dictionary
#make a plot from dictionary

nut_butter_prices={"Almond Butter":10,
                  "Peanut Butter":8,
                  "Cashew Butter":12}
fig,ax=plt.subplots(figsize=(5,4))
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)")
```

```python
## Horizontal bar creation,
fig,ax=plt.subplots(figsize=(5,4))
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))#It has to be turned into a list.
```

```python
# Plotting histogram,
#make some data
x=np.random.randn(1000)#The numpy.random.randn() function creates an array of specified shape and fills it with random values
fig,ax=plt.subplots(figsize=(5,4))
ax.hist(x)
```

## Importing csv file and performing visualizations

```python
df=pd.read_csv("car-sales.csv")
df
```

```python
# Replacing the expressions
df["Price"]=df["Price"].str.replace('$','')
df
```

```python
# Adding date column
df["Sale Date"]=pd.date_range("1/1/2020",periods=len(df))
df
```

```python
df.dtypes
```

```python
#Converting the price column to a numeric value. This example includes to keep the data with decimal points.
df["price"]=df["price"].str.replace('[\$\,]|\.\d*','').astype(float)
```

```python
df.dtypes
```

```python
#Appling Cumulative summation
df["Total Sales"]=df["price"].cumsum().astype(float)
df
```

```python
#lets plot the total/cumsum sales
#line plot
df.plot(x="Sale Date",y="Total Sales")
plt.ylabel("Price ($)")
```

```python
#reassign price column to int
df["price"]=df["price"].astype(int)
#scatter plot
df.plot(x="Odometer (KM)", y="price", kind="scatter")
```

```python
# Car graph from the imported data frame,
df.plot(x="Make", y="Odometer (KM)", kind="bar")
```

```python
#histogram
df["Odometer (KM)"].plot.hist()
#df["Odometer (KM)"].hist() #another way
```

```python
df["Total Sales"].plot(kind="bar",color=["salmon"])
```
